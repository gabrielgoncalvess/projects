# Biblioteca TextBlob

```
!pip install -U textblob
!python -m textblob.download_corpora
from textblob import TextBlob
```

```
def traduzir(frase):
  if TextBlob(frase).detect_language() != 'en':
      traducao = TextBlob(str(TextBlob(frase).translate(to='en')))
      #print('Tweet: {0} - Sentimento: {1}'.format(frase, traducao.sentiment))
      return str(traducao)
  else:
      return frase
```

```
tweets['tweet'] = tweets['tweet'].apply(lambda x: traduzir(x))
```
